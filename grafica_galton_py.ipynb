```python
import random
import matplotlib.pyplot as plot


#SIMULA LA CAIDA DE LAS CANICAS
def simula_canicas(num_canicas, num_niveles):

  """Simula la trayectoria de las canicas en la m치quina de Galton.
  Cada canida tiene igual probabilidad de caer a la izquierda o derecha
  en cada uno de los niveles

  Regresa la cantidad de canicas en cada contenedor.
  """

  #Inicializamos lista de ceros como contenedores posibles
  contenedores = [0] * (num_niveles +1 )

  for i in range(num_canicas):
    posicion = 0  #inicia en el centro, 0 saltos a la derecha

    for i in range(num_niveles):
      #movimiento aleatorio 0 =izquierda, 1= derechaa
      movimiento = random.choice([0, 1])
      if movimiento == 1:
        posicion += 1

    contenedores[posicion] += 1
  return contenedores



#GRAFICA RESULTADOS

def grafica_resultado(contenedores):
  """
  Genera una grafica de la distribuci칩n de canicas en los contenedores

  Par치metros:
  contenedores (lista):  cantidad de canicas en cada contenedor
  """

  num_contenedores = len(contenedores)
  indices = list(range(num_contenedores))

  plot.figure(figsize=((12, 6)))
  plot.bar(indices, contenedores, color='pink', edgecolor= 'black')
  plot.title("Simulacion de la M치quina de Galton")
  plot.xlabel("Numero de contenedor total de veces que se fue a la derecha)")
  plot.ylabel("Cantidad de canicas")
  plot.grid(axis= 'y', linestyle='--', alpha=1)
  plot.tight_layout()
  plot.show()


#PROGRAMA PRINCIPAL

if __name__ == "__main__":
      NUM_CANICAS = 3000
      NUM_NIVELES = 12

      #SIMULA Y GRAFICA
      resultados = simula_canicas(NUM_CANICAS, NUM_NIVELES)
      grafica_resultado(resultados)

```
